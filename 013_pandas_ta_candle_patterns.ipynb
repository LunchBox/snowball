## Pandas TA Candle Patterns

之前就有留意到 pandas_ta module 裡有一個 cdl_patterns，包含挺多著名的 K 線圖型態，仔細看了下，原來是封裝 TA-Lib 裡的 K線型態 patterns

封裝了也好，能夠一次過掃描所有的 pattern

```python
import pandas as pd
import pandas_ta as ta

import yfinance as yf
```

```python
df = yf.download("^SPX", period="1d", interval="5m")
```

```python
doji = df.ta.cdl_pattern(name="doji")

doji[doji["CDL_DOJI_10_0.1"] != 0]
```

將所有 candle pattern 一次過掃描出來

```python
import datetime
import mplfinance as mpf
```

```python
def mpf_show(data):
    position = data[data != 0]

    if len(position) != 0:
        print('position Pattern date:')
        print(position)
        for i in position.index[:3]: # 每種 pattern 畫個 3 幅圖就算
            st = i + datetime.timedelta(minutes=-10 * 5)
            et = i + datetime.timedelta(minutes=10 * 5)
            
            df_candle = df.loc[(df.index >= st) & (df.index <= et)]
            mpf.plot(df_candle, type='candle', title=i.strftime("%Y-%m-%d %H:%M"), figsize=(6,2))
            mpf.show()
            
```

mplfinance 畫 candle 圖挺好看的，數量太多，jupyter 也顯示不完

```python
for col in df.columns[6:]:
    print('------------------------------')
    print(col)
    mpf_show(df[col])
```
